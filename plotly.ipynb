```python
import numpy as np
A = np.array([
    [2,4,6],
    [8,10,121],
    [14,16,18]
])
B = np.array([
    [1,3,5],
    [7,9,11],
    [13,15,17]
])
print("Matrix A:\n",A)
print("\nMatrix B :\n",B)

print("shape of mat A\n",A.shape)
print("shape of mat B\n",B.shape)
print("Size of mat A\n",A.size)
print("Size of mat B\n",B.size)
print("dimension of mat A\n",A.ndim)
print("dimension of mat B\n",B.ndim)

```

```python
print("Addition of mat",A+B)
print("Substraction of mat",A-B)
print("matrix mul of a and b ",np.matmul(A,B))
print("transpose of A:",A.T)
print("transpose of B:",B.T)
print("det of A",np.linalg.det(A))
print("det of B",np.linalg.det(B))
print("inv of A",np.linalg.inv(A))
print("inv of B",np.linalg.inv(B))
print("trace of A",np.trace(A))
print("trace of A",np.trace(B))
print("max of mat A:",np.max(A))
print("min of mat A:",np.min(A))
print("mean of mat A:",np.mean(A))
print("sum of mat A:",np.sum(A))
```

```python
#row wise sum 
print("row wise sum")
print(np.sum(B, axis=1))
#col wise sum of B
print("\n col wise sum of B:")
print(np.sum(B, axis =0))
```

```python
#extract first row
print(A[0])
print("second col")
print(A[:,1])
print("last row")
print(A[-1])
print("bottom right elem")
print(A[-1,-1])
```

```python
#reshape mat A into a(1,9) arr
print("reshape a into (1,9)")
print(A.reshape(1,9))
print("flatter B")
print(B.flatten())
#eigenvalues, eigenvectors = np.linalg.eig(A)
print("eigenvalues")
print(np.linalg.eig(A))
print("eigenvector")
print(np.linalg.eig(A))
```

```python
# Horizontal
print("\nHorizontal Stack:")
print(np.hstack((A, B)))
# Vertical
print("\nVertical Stack:")
print(np.vstack((A, B)))
```
