```python
import numpy as np

# Student Marks Dataset (out of 100)
marks = np.array([45, 67, 89, 90, 56, 78, 88, 92, 61, 74,
                  83, 95, 52, 68, 71, 85, 77, 64, 58, 81])

print("Dataset:")
print(marks)
```

```python
print(np.mean(marks))
```

```python
print(np.median(marks))
```

```python
from scipy import stats
print(stats.mode(marks))
```

```python
print(np.max(marks))
```

```python
print(np.min(marks))
```

```python
print(np.var(marks))
```

```python
print(np.std(marks))
```

```python
print(np.percentile(marks,25))
```

```python
print(np.percentile(marks,50))
```

```python
print(np.percentile(marks,75))
```

```python
q1=np.percentile(marks,25)
q2 = np.percentile(marks,75)
print("interquartile range :",q1-q2)
```

```python
mean = np.mean(marks)
print("12. Students above mean:", np.sum(marks > mean))
```

```python
# 13. Students scoring below 60
print("13. Students below 60:", np.sum(marks < 60))
```

```python
total = np.sum(marks)
average = total / len(marks)
print("avg:",average)
print("total:",total)
```

```python
ascending = np.sort(marks)
descending = np.sort(marks)[::-1]
print("asc:",ascending)
print("desc:",descending)
```

```python
print("Top 5 Highest Scores:")
print(descending[:5])
```

```python
print(" Range:", np.max(marks) - np.min(marks))
```
