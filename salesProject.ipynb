```python
df["Sale_Date"] = pd.to_datetime(df["Sale_Date"])

monthly_sales = df.groupby(df["Sale_Date"].dt.to_period("M"))["Sales_Amount"].sum()

print(monthly_sales)
```

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
region_sales=df.groupby("Region",as_index=False)["Sales_Amount"].sum()

fig=px.bar(region_sales,
           x="Region",
           y="Sales_Amount",
           title="Region-wise Sales")

fig.show()
```

```python
category_sales=df.groupby("Product_Category",as_index=False)["Sales_Amount"].sum()

fig=px.pie(category_sales,
           names="Product_Category",
           values="Sales_Amount")

fig.show()
```

```python
monthly=df.groupby(df["Sale_Date"].dt.to_period("M"))["Sales_Amount"].sum().reset_index()

monthly["Sale_Date"]=monthly["Sale_Date"].astype(str)

fig=px.line(monthly,
            x="Sale_Date",
            y="Sales_Amount",
            markers=True)

fig.show()
```

```python
sales_rep=df.groupby("Sales_Rep",as_index=False)["Sales_Amount"].sum()

fig=px.bar(sales_rep,
           x="Sales_Rep",
           y="Sales_Amount")

fig.show()
```
